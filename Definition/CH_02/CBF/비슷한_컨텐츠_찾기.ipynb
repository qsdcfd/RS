### 유클리드 거리(Euclidean Distance)

image.png

- 두 점 사이의 거리를 계산할 때 사용한다.

- 여러 차원을 갖는 점과 점 사이의 거리를 계산한다.

- 거리 기반 유사도 측정 방식이다.

image.png

```python
from scipy.spatial import distance
a = (1, 2, 3)
b = (4, 5, 6)

print(distance.euclidean(a, b))
```

### 코사인 유사도-Cosine Similarity

image.png

- 벡터를 비교할 때, 가장 많이 유용하게 사용된다.

- 벡터A와 벡터B사이의 각도로 계산된다.

- -1과 1사이의 값을 가진다.

   - 완전히 반대:-1
   - 서로 독립:0
   - 완전히 같다:1
- 각도기반 유사도 측정 방법으로 두 벡터간 유사한 정도를 코사인 값으로 표현

image.png

image.png

```python
from scipy import spatial
List1 = [4, 47, 8, 3]
List2 = [3, 52, 12, 16]
result = 1 - spatial.distance.cosine(List1, List2)
print(result)
```
