### 콘텐츠 기반 추천시스템

**목적**

- 유저 A가 높은 평점을 주거나 크게 관심을 갖고 있는 아이템 X와 유사한 아이템 Y를 추천

<br>

**특징**

- 컨텐츠가 비슷한 아이템을 추천한다
- 사용자가 과거에 경험 혹은 구매했던 아이템 중에서 비슷한 아이템을 추천한다.
- 정보(아이템)을 찾는 과정과 과거 정보를 활용하여 유저의 성향을 배우는 문제이다.

image.png


<br>

**예시**

- 웹사이트,블로그,뉴스: 비슷한 컨텐츠의 게시글을 찾아서 추천
- 영화: 같은 배우, 같은 장르 같은 영화 감독 등 비슷한 특징을 갖는 영화를 찾아서 추천

<br>

**장점**

- 다른 유저의 데이터가 필요하지 않다.

- 추천할 수 있는 아이템 항목의 범위가 넓다.

   - Unique, New, Unpopular아이템 모두 가능

- 추천하는 이유를 제시한다.

   - 아이템의 features로 컨텐츠 분석 가능하다.
   - 특정 feature가 추천의 이유가 됐다고 설명 가능하다.

<br>

**단점**

- 다양한 features로 인해서 추천을 위한 적절한 features를 찾기가 어렵다

   - 영화, 사진, 음악, 뉴스(텍스트)등

- 새로운 유저를 위한 추천이 어렵다.

   - Cold start문제
   - User Profile가 존재하지 않거나 데이터가 매우 부족하다.

- 선호하는 특성을 가진 항목을 반복 추천한다.

   - 세부적인 파악이 되지 않고 큰 틀의 파악만 되기에 유저의 다양한 취한 반영이 어렵다
   - 유저 프로필 말고는 추천이 불가하다.
   - Overspcialization


<br>

**가장 중요한 것**

- 컨텐츠의 내용을 분석하는 아이템 분석 알고리즘

   - clustering, Machine Learning, TF-IDF등 화용
- 추천시스템의 성능을 높이는 적절한 콘텐츠 사용

   - 영화 추천을 위해서 영화 정보를 활용하고, 뉴스 추천을 위해서 뉴스 내용을 활용

<br>
