## 협업필터링이란?

**협업필터링(Collaborative Filtering)**

- 협동하여 필터링을 한다.

- 많은 사람의 의견을 바탕으로 더 나은 추천을 한다.

<br>

**집단지성이라고도 불린다**

- 개인보다는 단체 혹은 그룹의 선택과 취향에 의존한다.

- 여러 사람의 의견을 종합적으로 반영하고 다수의 의견을 통해서 더 나은 선택을 한다
