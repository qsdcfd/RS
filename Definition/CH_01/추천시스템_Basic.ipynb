## 추천시스템의 역사


### 추천시스템

**등장**

- 인터넷은 폭발적인 성장과 수많은 정보가 쏟아지는 공간이

- 인터넷관련 비즈니스 활성화

   -  e-Business, e-Commerce

- 수많은 정보와 데이터로 인해 사용자가 적절한 결정을 내리기 어려움

<br>

**정의**

- 사용자로부터 선호도 여부를 가지고 데이터화

   - Explicit Feedback(평점 등)
   - Implicit Feedback(사용자의 행동 패턴)

- 사용자의 선호도와 제한사항을 바탕으로 가장 적절한 아이템 점수 계산
- 사용자에게 가장 관련 있는 아이템 순서대로 리스트 형태로 제시

- TOP K를 잡는 것이 중요해졌다.

<br>

**사용 이유**

- 상업,기업입장에선 더 맣은 아이템을 판매할 수 있게 도와줌

- 소비자가 보지 못한 상품을 보여주어서 다양하게 판매할 수 있다

- 플랫폼이 잘 만들어지면, 소비자의 이용율과 접속율이 늘어나게 되기에 소비자 만족도가 올라간다.

- 충성고객이 증가한다.

-  데이터가 쌓이면 쌓일수록 알고리즘이 향상되기에 고객이 원하는 것이 무엇인지 알 수 있다.

<br>

**활용 예시**

- 넷플릭스: 콘텐츠 추천

   - 계정 생성시 선택된 '좋아하는 컨텐츠'를 기반으로 추천

<br>

- 인스타그램: 팔로우 추천

   - 팔로잉하고 있는 계정 유형과 피드에서 자주 클릭한 계정을 분석하여 피드 광고 노출

   - 계정 팔로우할 경우 해당 계정과 비슷한 계정을 추가적으로 추천

<br>

- 쿠팡: 상품 추천

   - 구매 목록, 상품 검색 키워드, 클릭한 상품을 기반으로 다양한 상품 추천

<br>

**데이터 종류**

- items

   - 추천할 항목

   - 가치와 데이터의 복잡함에 따라 구분
   
   - Low: 뉴스, 책, 영화

   - High : 금융 상품, 직업, 여행


- Users

   - 여러 목적과 특징을 갖는 집단

   - 행동 패턴등으로 유저 모델링

   - 협업 필터링: 여러 아이템에 대한 평점 리스트

   - 사회적인 요인(유행, 분위기)을 활용한 추천시스템

- Transactions

   - 유저와 아이템간의 상호 작용

   - 유저가 인터넷에 남긴 로그에서 중요한 정보 추출
   - 해시태그, 평점, implicit한 정보 모두 활용 가능

### 추천시스템이 풀고 싶은 문제

**랭킹 문제**

- 특정 유저가 특정 아이템에 대한 평점 or 점수를 정확하게 예측할 필요 없음

- 특정 아이템을 좋아할만한 Top-K유저를 선정할 수 있음

- 특정 유저가 좋아할만한 Top-K아이템을 선정할 수 있음

<br>

**예측 문제**

- 유저-아이템 조합에서 평점 or 점수 예측

- 유저-아이템 행렬을 채우는 문제

- 유저(m명)-아이템(n)개의 mXn행렬, 그러나 비어있는 부분 존재

   - 관측값(Observed value)은 모델 학습에 사용
   - 결측값(Missing value)는 모델 평가에 사용
