## 추천알고리즘

### 컨텐츠기반 추천시스템(Contents-based Recommender System)

- 사용자가 과거에 좋아했던 아이템을 파악하고, 그 아이템과 비슷한 아이템을 추천

- 예시:)

   - 영화 스파이더맨에 4.5점 평점을 부여한 유저는 타이타닉보다 캡틴 마블을 더 좋아할 것이다.

### Collaborative Filtering(협업필터링)

- 비슷한 성향 또는 취향을 갖는 다른 유저가 좋아한 아이템을 현재 유저에게 추천한다.

- 간단하지만 수준 높은 정확도를 나타낸다.

- 예시:)

   - 스파이더맨에 4.5점을 준 2명의 유저에게 유저 A가 과거 좋아했던 캡틴 마블을 유저 B에게 추천한다.

### Hybrid Recommender System

- Content-based와 Collaborative Filtering의 장단점을 상호보안

- Collaborative Filtering은 새로운 아이템에 대한 추천 부족으로 인해서 content-based기법이 cold start문제에 도움을 준다.

- cold start

   - 협업 필터링 알고리즘을 사용하기 위해서 어느정도 기존 데이터 필요

   - 신규 사용자에겐 어떤 아이템도 추천할 수 없다.

   - 이러한 상황('새로 시작할 때의 곤란함')을 cold start라고한다.

#### 추가적인 추천 알고리즘 소개

**Context-based Recommendation**

   - Context-aware Recommendation System

   - Location-based Recommendation System

   - Real-time or Time-Sensitive Recommendation Syystem

<br>

**Community-based Recommendation**

- 사용자의 친구 또는 속한 커뮤니티의 선호도를 바탕으로 추천

- SNS등의 뉴스피드 또는 SNS네트워크 데이터 활용

<br>

**Knowledge-based Recommendation**

- 특정 도메인 지식을 바탕으로 아이템의 features를 활용한 추천

- Case-based Recommendation

   - 사용자의 니즈와 해결책 중 가장 적합한 것을 골라서 추천

- Constraint-based Recommendation

   - 사용자에게 추천할 때, 정해진 규칙을 바탕으로 추천
